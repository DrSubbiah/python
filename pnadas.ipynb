We can also filter out the string columns in the data by using **str.contains()**. 

And adding a "~" to the beginning of the line creates the opposite result.

```python
df.loc[df['abilities'].str.contains('Solar')]
```

The **groupby()** function is another powerful tool for filtering. This sort of filtering can be done in a single line instead of multiple lines of code and loops.

You can also add a **mutate()** or **map()** function to the end to alter the dataset to your needs.

```python
df1 = df.groupby(['type1'], as_index = False).mean()
df1
```

```python
df2 = df.groupby('type1', as_index = False)['hp'].mean()
df2
```

```python
df2.plot(x='type1', y= "hp", kind="bar")
```

```python
df3 = df.loc[df['is_legendary'] == 1]
df4 = df3.groupby('type1', as_index = False)['is_legendary'].count()
df4
df4.plot(x='type1', y= "is_legendary", kind="bar")
```

```python
#df5 = df[['generation', 'type1','sp_attack']]
df5 = df.groupby(['type1','generation'],as_index=False)['sp_attack'].mean()
print(df5.describe())
df5= df5.astype({'generation': str})
df5['new'] = df5['type1'] + df5['generation']
print(df5)
```

**You can also do groupby() to multiple columns as shown above.**

```python
df5.plot(x='type1', y= "sp_attack", kind="bar")
```

From the above graph we can infer that a newer generation doesnt mean a better attack power.
